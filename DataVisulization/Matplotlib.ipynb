```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Matplotlib Basic

```python
x = np.linspace(1,10,20)
y = x**2
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

## Subplot in Matplotlib 

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(np.sin(x), np.sin(y))
plt.subplot(2,2,3)
plt.plot(np.cos(x),np.cos(y))
plt.subplot(2,2,4)
plt.plot(np.tan(x),np.tan(y))
```

## Object Oriented way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(13,8))
canvas1 = fig.add_axes([0.1,0.1,1,1])
canvas1.plot(x,y)
canvas2 = fig.add_axes([0.1,0.8,0.4,0.3])
canvas2.plot(x,y)
```

## Types of plots

## Saving Canvas

```python
fig = plt.figure()
canvas = fig.add_axes([0.1,0.1,0.8,0.8])
canvas.plot(x,y)
plt.savefig("BasicPlot.png")
```
