```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset("flights")
tips =  sns.load_dataset("tips")
```

## Heat Map

```python
tipscorr  = tips[['total_bill','tip','size']]
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

## Cluster Map

```python
sns.clustermap(tipscorr.corr(),annot=True)
```

## Pivot Table Heat Map

```python
flights
```

```python
pvtFlights = flights.pivot_table(values='passengers',columns="month",index="year")
pvtFlights
```
