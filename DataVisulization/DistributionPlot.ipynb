```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

*  distplot/histplot
*  jointplot
*  pairplot
*  rugplot

```python
df = sns.load_dataset("tips")
df.head()
```

## Hist Plot

```python
plt.subplot(2,2,1)
sns.histplot(df['total_bill'])
plt.subplot(2,2,2)
sns.histplot(df['total_bill'],bins=20)
plt.subplot(2,2,3)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(2,2,4)
sns.histplot(df['tip'],bins=20,kde=True)
```

## Joint Plot

```python
# The all value of kind is (scatter,kde,hist,hex,reg,reside) and (hex,reg,reside) are not work when you use hue
sns.jointplot(df,x="total_bill",y="tip",kind="scatter",hue="sex")
```

## Pair Plot

```python
sns.pairplot(df,hue="sex")
```

## Rug Plot

```python
sns.rugplot(df)
```
