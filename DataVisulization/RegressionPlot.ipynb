```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## lmplot()

```python
tip = sns.load_dataset("tips")
```

The values of markers is 
  * 'o': 'circle',
  * 's': 'square',
  * 'D': 'diamond',
  * 'd': 'thin_diamond',
  * '^': 'triangle_up',
  * 'v': 'triangle_down',
  * '<': 'triangle_left',
  * '>': 'triangle_right',
  * 'p': 'pentagon',
  * 'h': 'hexagon',
  * 'H': 'rotated_hexagon',
  * '8': 'octagon',
  * '+': 'plus',
  * 'x': 'x',
  * 'X': 'filled_x',
  * 'P': 'filled_plus',
  * '*': 'star',
  * '.': 'point',
  * ',': 'pixel',
  * '_': 'horizontal_line',
  * '|': 'vertical_line'


```python
sns.lmplot(data=tip,x="tip",y="total_bill",hue="sex",markers=["v","o"],scatter_kws={"s":30})
```
